```python
# import important modules 
import zipfile
import xml.etree.ElementTree as ET
from bs4 import BeautifulSoup
```

```python
# unzip the file by using zipfile moduel:
filename = 'D:\\python\\xml_containing_html.xml.zip' # xml file on the local computer


with zipfile.ZipFile(filename, 'r') as zip_ref:
    zip_ref.extractall('D:\\python')
    
 # parse xml file by using ElementTree:   
tree = ET.parse('D:\\python\\xml_containing_html.xml')
root = tree.getroot()
```

```python
# Created a dictionary to extract html, parsed it and put it together.
thisdict={}
for child in root: 
    thisdict.update({child.tag:child[0].text})

```

```python
#extracted htm files
thisdict
```

```python
# paresed the html file using BeautifulSoup moduel and get all href links, then extracted the required links from each page and count total links:

totalLinks=0
pagenumber=-1
pagecount=[*thisdict]
for key in thisdict:
    soup = BeautifulSoup(thisdict[key],'html.parser')
    pagenumber=pagenumber+1
    linkcount=0
    
    for link in soup.find_all('a', href=True):
        linkcount=linkcount+1
    totalLinks=totalLinks+linkcount
    print('The number of links with tag <a href> on page',pagenumber+1, 'with',pagecount[pagenumber], ' =', linkcount)
print('Total Links with tag <a href> =', totalLinks)
```
